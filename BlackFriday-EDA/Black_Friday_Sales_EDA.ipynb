# <b>Black Friday Dataset EDA And Feature Engineering

## Cleaning and preparing the data for model training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# <b>Problem Statement</b>

A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
#importing the dataset
df_train=pd.read_csv('train.csv')
df_train.head()
```

```python
# import the test Data
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
# Concat both trin and test data
#The ignore_index=True argument is optional but helps to reindex the resulting data frame.
#df = df_train.append(df_test, ignore_index=True)   # Not working. why?
df = pd.concat([df_train, df_test], ignore_index=True)
df.head()
```

```python
# Basic
df.info()
```

```python
df.describe()
```

```python
# here user ID has no role to play. so delete it.
# axis=0 means drop it horizonatly or row wise
# axis=1 means drop it vertically or column wise
# inplace=True means it will automatically update into the df value
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

## Next two codes are used for Handling Categorical feature Gender

```python
# convert any categorical feature to numeric value
#pd.get_dummies(df['Gender'])
#pd.get_dummies(df['Gender'],drop_first=1)

# for directly assigning 0 and 1 value in Dataset see next code IMP.
# OR given below code can also be used.
df['Gender']=pd.get_dummies(df['Gender'], drop_first=1).astype(int)
df.head()
```

```python
# here it is done with the help of mapping
#df['Gender']=df['Gender'].map({'F': 0,'M': 1})
#df.head()
```

```python
# Handle categorical feature Age
df['Age'].unique()
```

```python
# not good practice
pd.get_dummies(df['Age'],drop_first=True)
```

```python
# Target Guiding or Target ordinal Encoding.
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
#Label encoding can also be done but not prefered here.
```

```python
# fixing categorical City_category
df_city=pd.get_dummies(df['City_Category'],drop_first=True).astype(int)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
# drop the city category feature
df.drop('City_Category',axis=1,inplace=True)
df.head()
```

```python
# Missing values
df.isnull().sum()
```

```python
# Focus on replacing missing values
df['Product_Category_2'].unique()
# it is dicrete feature
```

```python
df['Product_Category_2'].value_counts()
```

```python
# Replace the missing value with the mode
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'])
```

```python
df['Product_Category_2'].mode()[0]
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
# product_category 3
df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'].value_counts()
```

```python
# Replace the missing value with the mode
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3']).mode()[0]
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
df.info()
```

```python
# convert object into integers
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
df.info()
```
